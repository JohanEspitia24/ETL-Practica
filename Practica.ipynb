```python
import os
from os import path
import shutil

def listing(mypath):
    items = [os.path.join(mypath, f) for f in os.listdir(mypath)]
    for item in items:
        print(item)

#Create a folder
print('Create a folder')
dir_path = 'C:/Users/johan/Documents/new_folder'

try:
    if not path.exists(dir_path):
        os.makedirs(dir_path)

except PermissionError:
    print('Acceso denegado por permisos de administrador')
listing('C:/Users/johan/Documents')

#Copy file
print('Copy file')
source = 'C:/Users/johan/Documents/archive/Renewable_Energy.csv'
destination = 'C:/Users/johan/Documents/new_folder/Renewable_Energy.csv'

shutil.copy(source, destination) # copy the file 'Renewable_Energy.csv' to the folder 'new_folder'
listing('C:/Users/johan/Documents/new_folder')

#Rename file
print('Rename file')
os.rename('C:/Users/johan/Documents/new_folder/Renewable_Energy.csv', 'C:/Users/johan/Documents/new_folder/Rename_Energy.csv') # rename the file 'Renewable_Energy.csv' to 'Renewable_Energy_2.csv'
listing('C:/Users/johan/Documents/new_folder')

#Move file
print('Move file')
shutil.move('C:/Users/johan/Dowloads/archive/Renewable_Energy.geojson', 'C:/Users/johan/Documents/new_folder') # move the file 'Renewable_Energy_2.csv' to the folder 'Documents'
listing('C:/Users/johan/Documents/new_folder')

```
