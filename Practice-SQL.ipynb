```python
import sqlite3 
sqlitedb = 'demo.db'

#Connect db
conn = sqlite3.connect(sqlitedb)
conn
```

```python
#CRUD - Create, Read, Update, Delete
cursor = conn.cursor()
```

```python
#Crate table
sql = '''CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name char(50) NOT NULL,
    age int,
    email char(50)
)'''

cursor.execute(sql)
conn.commit()
print('Table created successfully')
```

```python
#Read all the data
cursor.execute("SELECT * FROM users" )
result = cursor.fetchall()
conn.commit()
result
```

```python
#Read only one data
cursor.execute("SELECT * FROM users" )
result = cursor.fetchone()
conn.commit()
result
```

```python
#Read a specific data from the table
params = (25,)
cursor.execute("SELECT * FROM users WHERE age > ?", params)
result = cursor.fetchall()
conn.commit()
result
```

```python
#Update data
params = (31, "johan1234@symphony")
cursor.execute(
    "UPDATE users SET age = ?, email = ? ", params
) 
conn.commit()
print('Data updated successfully')

#Read table
cursor.execute("SELECT * FROM users" )
result = cursor.fetchall()
conn.commit()
result
```

```python
#Delete data
params = (31,)
cursor.execute("DELETE FROM users WHERE age = ?", params)
conn.commit()
result = cursor.fetchall()
result

cursor.execute("SELECT * FROM users" )
conn.commit()
result = cursor.fetchall()
result
```

```python
#Query with pandas
import pandas as pd

df = pd.read_sql_query("SELECT * FROM users", conn)
df
```

```python
#Dataframe to table
products = {
    'ID': ['emp001', 'emp002'],
    'Name': ['John Doe', 'Jane Smith'],
    'Age': [25, 30],
    'Email': ['johndoe@example.com', 'janes@example.com']
    }

dataFrame = pd.DataFrame(products)
dataFrame    
```

```python
#Save dataframe from sqlite
tableName = 'employees'
dataFrame.to_sql(tableName, con = conn, if_exists='append', index=False)

#Check
df = pd.read_sql_query("SELECT * FROM employees", conn)
df
```
